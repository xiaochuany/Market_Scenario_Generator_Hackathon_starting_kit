```python
import pickle
from pathlib import Path

import numpy as np
import polars as pl
```

```python
path_data = Path("data/ref_data.pkl")
path_label = Path("data/ref_label.pkl")

with open(path_data, "rb") as f1, open(path_label, "rb") as f2:
    data = pickle.load(f1)
    label = pickle.load(f2)
```

```python
def visual(data,label,idx):
    return pl.DataFrame(data[idx].T).plot(label=f"crisis = {label[idx].item()}")
```

```python
def get_crisis_idx(label):
    idx_arr = label.argsort(0).reshape(-1)[-label.sum().astype(np.int32):]
    idx_arr.sort()
    return idx_arr

get_crisis_idx(label)[-10:]
```

```python
visual(data,label,1087)
```
